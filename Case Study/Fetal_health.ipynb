# feature selection

```python
data.columns
```

```python
# all the feature can be selected due to its correlanility
X = data.drop(columns = 'fetal_health')
Y = data['fetal_health']
```

# splitting the training and testing data

```python
from sklearn.model_selection import train_test_split
train1, test1, train2, test2 = train_test_split(X, Y, test_size = 0.25)
```

```python
# standardization of data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train1)

X_train = scaler.transform(train1)
X_test = scaler.transform(test1)
```

# SVM Model

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='rbf', random_state = 1)
svm_model.fit(train1, train2)
```

```python
svm_pred = svm_model.predict(test1)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
random_model = RandomForestClassifier()
```

```python
random_model.fit(train1, train2)
random_pred = random_model.predict(test1)
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
decision_model = DecisionTreeClassifier()
```

```python
decision_model.fit(train1, train2)
decision_pred = decision_model.predict(test1)
```

# KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(train1, train2)
```

```python
knn_pred = knn_model.predict(test1)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print('accuracy: ', accuracy_score(svm_pred, test2))
print('classification report \n', classification_report(svm_pred, test2))
print('\n confusion matrix', confusion_matrix(svm_pred, test2))
```

```python
# Random Forest Model Evaluation
print('accuracy: ', accuracy_score(decision_pred, test2))
print('classification report \n', classification_report(decision_pred, test2))
print('\n confusion matrix', confusion_matrix(decision_pred, test2))
```

```python
print('accuracy: ', accuracy_score(random_pred, test2))
print('classification report \n', classification_report(random_pred, test2))
print('\n confusion matrix', confusion_matrix(random_pred, test2))
```

```python
# knn model evaluation
print('accuracy: ', accuracy_score(knn_pred, test2))
print('classification report \n', classification_report(knn_pred, test2))
print('\n confusion matrix', confusion_matrix(knn_pred, test2))
```
